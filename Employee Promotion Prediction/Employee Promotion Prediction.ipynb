## Missing values

```python
# Create a datframe of missing values in both datasets
train_missing = train.isna().sum()
test_missing = test.isna().sum()

train_missing_perc = train_missing.div(train.shape[0]) * 100
test_missing_perc = test_missing.div(test.shape[0]) * 100

missing_values = pd.concat(objs = [train_missing, train_missing_perc, test_missing, test_missing_perc],
                      axis = 1, keys = ['train_missing','train%', 'test_missing', 'test%'])
missing_values
```

There is only 2 columns with missing values - education and previous year rating. Percentage of missing values is quiete low, we don't have to delete any values. 

## Impute missing values

```python
# Fill missing values in both datasets with mode
train.education.fillna(value = train.education.mode()[0], inplace = True)
train.previous_year_rating.fillna(value = train.previous_year_rating.mode()[0], inplace = True)

test.education.fillna(value = test.education.mode()[0], inplace = True)
test.previous_year_rating.fillna(value = test.previous_year_rating.mode()[0], inplace = True)
```

```python
# Checking if there is still any missing values
train_missing = train.isna().sum().sum()
test_missing = train.isna().sum().sum()
print(f'Missing values in train set: {train_missing}')
print(f'Missing values in test set: {test_missing}')
```

## Outliers

```python
# Check numerical columns in dataframe
train.select_dtypes('number').head()
```

```python
# We suspect previous_year_rating and length_of_service to have outliersb
prob_outliers = ['previous_year_rating', 'length_of_service']
# We can easily check the existance of outliers using boxplots
for column_name in prob_outliers:
    fig = px.box(data_frame = train, x = column_name)
    fig.show(renderer="svg")
```

## Feature engineering

```python
# We can boost performance of machine learning model by creating extra columns from already existing ones.
train['total_score'] = train.no_of_trainings * train.avg_training_score
test['total_score'] = test.no_of_trainings * test.avg_training_score
```

```python
# Deleting columns that won't add any value to machine learning model
train.drop(columns = ['employee_id', 'region','recruitment_channel'], inplace = True)
test.drop(columns = ['employee_id', 'region','recruitment_channel'], inplace = True)
```

## Categorical encoding

```python
# To train model we need to change all the object type columns to numeric values
train.select_dtypes('object').head()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
# We can encode object type columns using LabelEncoder model
le = LabelEncoder()

# Transforming object type columns from both of the dataframes to numeric
train.department = le.fit_transform(train.department)
train.gender = le.fit_transform(train.gender)

test.department = le.fit_transform(test.department)
test.gender = le.fit_transform(test.gender)
```

```python
# Manually changing columns with more classes than 2, by mapping them to numeric values
train.education = train.education.map({"Master's & above": 2, "Bachelor's": 1, "Below Secondary": 0})
test.education = test.education.map({"Master's & above": 2, "Bachelor's": 1, "Below Secondary": 0})
```

## Splitting data

```python
# Splitting train set to x and y
y = train.is_promoted
x = train.drop('is_promoted', axis = 1, inplace = False)
```

```python
# Checking if shape of sliced dataframe is correct
print(f'Shape of x is: {x.shape}')
print(f'Shape of y is: {y.shape}')
print(f'Shape of test is: {test.shape}')
```

## Resampling

```python
from imblearn.over_sampling import SMOTE
```

```python
# Solving target class disproportion problem using resamping
x_resample, y_resample = SMOTE().fit_resample(x, y.values.ravel())
```

## Normalization

```python
from sklearn.preprocessing import StandardScaler
```

```python
# Normalize dataset using StandardScaler
sc = StandardScaler()
X = sc.fit_transform(x_resample)
```

## Train Test split

```python
from sklearn.model_selection import train_test_split
```

```python
# Splitting data into train and validation datasets
x_train, x_test, y_train, y_test = train_test_split(X, y_resample, test_size = 0.2)
```

## Machine learning model - Decision Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
# Initialize model, training and getting predictions based on validation samples
model = DecisionTreeClassifier()
model.fit(x_train, y_train)
predictions = model.predict(x_test)
```

## Performacje analysis

```python
# Checking accuracy score of our model
print(f'Training accuracy: {model.score(x_train, y_train)}')
print(f'Validation accuracy: {model.score(x_test, y_test)}')
```

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
# Plotting confusion matrix to check performance of our model
cm = confusion_matrix(y_true = y_test, y_pred = predictions)
cm_df = pd.DataFrame(data = cm)
fig = px.imshow(cm, text_auto = True)
fig.update_xaxes(title_text='Predicted')
fig.update_yaxes(title_text='True')
fig.show(renderer="svg")
```

```python
# Decision tree trained model info
report = classification_report(y_true = y_test, y_pred = predictions)
print(report)
```

__Q2. What is the Probability to get Promoted, If an employeed has won an award?__

```python
award_mask = train['awards_won?'] == 1
award_df = train.loc[award_mask]
not_promoted, promoted = award_df.is_promoted.value_counts().values
probabilty = promoted / (not_promoted + promoted)
print(f'if employee has won an award, his chances to get promoteted are: {round(probabilty,2) * 100}%')
```

__Q3. What is the Average Training Score of those Employees who got Promotion?__

```python
promotion_mask = train.is_promoted == 1
promotion_df = train.loc[promotion_mask]
avg_score = promotion_df.avg_training_score.mean()
print(f'Average training score for promoted employees is: {round(avg_score,2)} points.')
```

__Q4. What is the Impact of Gender in Promotions?__

```python
corr_df = train.corr()
corr = corr_df.loc['gender','is_promoted']
print(f'Impact of gender in promotions is equals to: {round(corr,3)}')
```

__Q5. What is the Probability of Freshers getting Promoted?__

```python
fresher_age = 30
fresher_service = 2

age_mask = train.age <= fresher_age
service_mask = train.length_of_service <= fresher_service
freshers_df = train.loc[age_mask & service_mask]
not_promoted, promoted = freshers_df.value_counts('is_promoted').values
prob = promoted / (not_promoted + promoted)
print(f'Probability of freshers getting promoted is: {round(prob,2) * 100}%')
```
